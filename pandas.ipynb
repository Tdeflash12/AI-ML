```python
import numpy as np
import pandas as pd
```

```python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka,Bhutan']

pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]

pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
# setting a game
marks=pd.Series(marks,index=subjects,name="Abhesh Marks")
```
